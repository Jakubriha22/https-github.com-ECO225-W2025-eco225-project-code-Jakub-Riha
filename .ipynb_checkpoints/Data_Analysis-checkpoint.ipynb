```python
# Import necessary libraries
import pandas as pd
import statsmodels.api as sm

# Define file path (Make sure this matches where you saved the dataset)
file_path = "/Users/jakubriha/Desktop/ECO225/Project/Data/data_df.csv"  

# Load the dataset
data_df = pd.read_csv(file_path)
```

```python
# Ensure numeric types where applicable
data_df['gender_dummy'] = data_df['gender_dummy'].astype(int)
data_df['loan_amount'] = data_df['loan_amount'].astype(int)  # If necessary, ensure it's an integer
data_df['hdi'] = data_df['hdi'].astype(float)  # HDI is typically a float
data_df['term_in_months'] = data_df['term_in_months'].astype(int)
data_df['funding_time_days'] = data_df['funding_time_days'].astype(int)
data_df['log_funding_time'] = data_df['log_funding_time'].astype(float)  # Log-transformed values are usually floats
```

```python
def run_robust_regression(y, X, model_name):
    """Runs OLS regression with robust standard errors and prints the summary."""
    X = sm.add_constant(X)  # Add intercept
    model = sm.OLS(y, X, missing='drop').fit(cov_type='HC0')  # HC0 for robust SE
    print(f"\nRegression: {model_name}")
    print(model.summary())

# Define dependent variables
funding_time_y = data_df['funding_time_days']  # Raw funding time
log_funding_time_y = data_df['log_funding_time']  # Log-transformed funding time

# Define independent variables
X_female = data_df[['gender_dummy']].astype(int)
X_loan_amount = data_df[['loan_amount']].astype(int)
X_hdi = data_df[['hdi']].astype(float)  # HDI should remain a float
X_loan_terms = data_df[['term_in_months']].astype(int)

# Create interaction term (ensuring correct type)
data_df['hdi_female_interaction'] = (data_df['hdi'] * data_df['gender_dummy']).astype(float)

# Extract sector fixed effects (dummy variables) and ensure they are integers
sector_dummies = pd.get_dummies(data_df['sector_name'], drop_first=True).astype(int)
X_sector_fixed = sector_dummies

# Multiple independent variable sets (ensuring correct type)
X_hdi_female = data_df[['hdi', 'gender_dummy', 'hdi_female_interaction']].astype(float)
X_full = pd.concat([data_df[['gender_dummy', 'loan_amount', 'hdi', 'term_in_months', 'hdi_female_interaction']], 
                    X_sector_fixed], axis=1).astype(int)
```

```python
# Run all regressions (Raw Funding Time)
print("\n=== Regressions with Raw Funding Time ===")
run_robust_regression(funding_time_y, X_female, "Funding Time ~ Female")
run_robust_regression(funding_time_y, X_loan_amount, "Funding Time ~ Loan Amount")
run_robust_regression(funding_time_y, X_hdi, "Funding Time ~ HDI")
run_robust_regression(funding_time_y, X_loan_terms, "Funding Time ~ Loan Terms")
run_robust_regression(funding_time_y, X_sector_fixed, "Funding Time ~ Sector Fixed Effects")
run_robust_regression(funding_time_y, X_hdi_female, "Funding Time ~ HDI + Female + HDI*Female")
run_robust_regression(funding_time_y, X_full, "Funding Time ~ Female + Loan Amount + HDI + Loan Terms + Sector Fixed Effects + HDI*Female")
```

```python
# Run all regressions (Log Funding Time)
print("\n=== Regressions with Log Funding Time ===")
run_robust_regression(log_funding_time_y, X_female, "Log Funding Time ~ Female")
run_robust_regression(log_funding_time_y, X_loan_amount, "Log Funding Time ~ Loan Amount")
run_robust_regression(log_funding_time_y, X_hdi, "Log Funding Time ~ HDI")
run_robust_regression(log_funding_time_y, X_loan_terms, "Log Funding Time ~ Loan Terms")
run_robust_regression(log_funding_time_y, X_sector_fixed, "Log Funding Time ~ Sector Fixed Effects")
run_robust_regression(log_funding_time_y, X_hdi_female, "Log Funding Time ~ HDI + Female + HDI*Female")
run_robust_regression(log_funding_time_y, X_full, "Log Funding Time ~ Female + Loan Amount + HDI + Loan Terms + Sector Fixed Effects + HDI*Female")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Bin Loan Amount and Loan Term into categories
data_df["loan_amount_bin"] = pd.qcut(data_df["loan_amount"], q=10, labels=[f"Q{i+1}" for i in range(10)], duplicates="drop")
loan_term_bins = np.linspace(data_df["term_in_months"].min(), data_df["term_in_months"].max(), 11)
loan_term_labels = [f"{int(loan_term_bins[i])}-{int(loan_term_bins[i+1])}" for i in range(len(loan_term_bins)-1)]
data_df["loan_term_bin"] = pd.cut(data_df["term_in_months"], bins=loan_term_bins, labels=loan_term_labels)

# Compute average funding time for each Loan Amount - Loan Term combination
heatmap_data = data_df.groupby(["loan_amount_bin", "loan_term_bin"])["funding_time_days"].mean().unstack()

# Replace missing values (no observations) with NaN (which will be mapped to white)
heatmap_data = heatmap_data.fillna(np.nan)

# Create a custom colormap where high values are red and low values are blue
cmap = sns.color_palette("coolwarm", as_cmap=True)  # No "_r", so red is for high values

# Create heatmap
plt.figure(figsize=(10, 6))
ax = sns.heatmap(
    heatmap_data, cmap=cmap, annot=True, fmt=".1f", linewidths=0.5,
    mask=heatmap_data.isna(), cbar_kws={'label': 'Average Funding Time (Days)'}
)

# Set axis labels and title with clear descriptions
plt.xlabel("Loan Term (Month Ranges)", fontsize=12)
plt.ylabel("Loan Amount (Deciles of Full Range)", fontsize=12)
plt.title("Heatmap of Average Funding Time by Loan Amount and Loan Term", fontsize=14)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha="right")

# Show plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Check if 'sector_name' is already a column (not an index)
if 'sector_name' not in data_df.columns:
    data_df = data_df.reset_index()

# Bin Loan Term into equal-width bins for better visualization
loan_term_bins = np.linspace(data_df['term_in_months'].min(), data_df['term_in_months'].max(), 11)
loan_term_labels = [f"{int(loan_term_bins[i])}-{int(loan_term_bins[i+1])}" for i in range(len(loan_term_bins)-1)]
data_df['Loan_Term_Binned'] = pd.cut(data_df['term_in_months'], bins=loan_term_bins, labels=loan_term_labels, include_lowest=True)

# Compute the average funding time per sector and loan term
heatmap_data = data_df.groupby(['sector_name', 'Loan_Term_Binned'])['funding_time_days'].mean().unstack()

# Sort sectors by the number of missing values (to move missing data to the top)
heatmap_data = heatmap_data.loc[heatmap_data.isnull().sum(axis=1).sort_values(ascending=False).index]

# Replace missing values (NaN) with white cells
mask = heatmap_data.isna()

# **Find the center of the colormap for better balance**
center_value = heatmap_data.mean().mean()  # Mean funding time as center

# Plot heatmap with balanced color palette
plt.figure(figsize=(12, 6))
sns.heatmap(
    heatmap_data, 
    cmap="coolwarm", 
    annot=True, 
    fmt=".1f", 
    linewidths=0.5, 
    mask=mask, 
    center=center_value,  # Center colormap around the mean
    cbar_kws={'label': 'Average Funding Time (Days)'}
)

# Set axis labels and title with clear descriptions
plt.xlabel("Loan Term (Month Ranges)", fontsize=12)
plt.ylabel("Sector", fontsize=12)
plt.title("Heatmap of Average Funding Time by Loan Term and Sector", fontsize=14)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha="right")

# Show plot
plt.show()
```
