# What Drives Loan Funding Speed? Evidence from Loan Terms, Size, and Borrower Characteristics

# 1. Introduction

In microfinance, the speed at which loans are funded plays a crucial role in determining the success and sustainability of both borrowers and lending institutions. Understanding the key funding speed drivers can offer insights into improving microfinance efficiency and equity. Although microloans are often disbursed prior to full funding, the speed at which they reach full funding can reflect lender preferences, signal borrower creditworthiness, and influence the availability of future lending opportunities. This project investigates the impact of loan terms, loan size, and borrower characteristics such as gender, geographic location, and sector, on the time it takes for loans to be fully funded.

The dataset for this analysis comes from Kiva.org, a microfinance platform that connects lenders with borrowers worldwide. It provides information on loan attributes, borrower demographics, and funding timelines. Our research focuses on identifying which factors most significantly influence funding speed.

Existing literature offers valuable foundations for this investigation. Pham and Shen (2017) explore how group loan structures affect microfinance outcomes, identifying borrower dynamics that may influence funding efficiency. Hermes and Hudon (2018) provide a systematic review of microfinance performance determinants, emphasising institutional and loan-specific variables relevant to funding speed. Additionally, Rauf et al. (2022) analyse how macro and micro factors shape microfinance interest rates, indirectly highlighting the role of loan terms and sizes in microloan decision-making. Kiros (2022) and Arene (1992) contribute insights into borrower behaviour and loan performance, which are critical when examining how borrower characteristics impact funding timelines.

Preliminary findings from this project suggest that loan sector plays a pivotal role in determining funding speed, with certain sectors consistently receiving faster funding. Furthermore, larger loan amounts and longer loan terms tend to experience delays in funding. Interestingly, gender also emerges as a significant factor; female borrowers are generally funded faster than male borrowers, with further intersectoral gender variance. However, there currently appears to be no significant relationship between HDI of the borrower's country and funding speed.

This project aims to fill gaps in the existing literature by providing empirical evidence on the relationship between loan-specific and borrower-specific factors in determining funding speed. Through empirical data analysis, we aim to offer practical implications for both lenders and borrowers.

# 2. Data Cleaning & loading

```python
# Importing Libraries
import pandas as pd
import numpy as np
import os
import pycountry
import matplotlib.pyplot as plt
import seaborn as sns

# File Paths
data_path = "/Users/jakubriha/Desktop/ECO225/Project/Data/"
loans_file = data_path + "loans.csv"
kiva_loans_file = data_path + "kiva_loans.csv"
country_stats_file = data_path + "country_stats.csv"
loan_coords_file = data_path + "loan_coords.csv"
```

```python
# Loading and Merging Data

df1 = pd.read_csv(loans_file)
df2 = pd.read_csv(kiva_loans_file).rename(columns={'id': 'loan_id'})
merged_df = df1.merge(df2, on='loan_id', how='inner')
```

```python
# Cleaning Inconsistent and Empty Observervations

# Drop rows with missing values in key columns
required_columns = ['funded_amount_x', 'loan_amount_x', 'country_code_x',
                    'borrower_genders_x']
filtered_df = merged_df.dropna(subset=required_columns).copy()

# Remove discrepancies in categorical columns
columns_to_check = ['country_code', 'currency', 'borrower_genders', 
                    'repayment_interval']
for col in columns_to_check:
    filtered_df = filtered_df[filtered_df[f'{col}_x'] == 
    filtered_df[f'{col}_y']]

# Keep only one version of duplicated columns
filtered_df = filtered_df.rename(columns={f'{col}_x': col for col in 
                                          columns_to_check})
filtered_df.drop(columns=[f'{col}_y' for col in columns_to_check], 
                 inplace=True)

# Remove all variables with '_y' and rename corresponding '_x' columns
columns_to_remove = [col for col in filtered_df.columns if 
                     col.endswith('_y')]
filtered_df.drop(columns=columns_to_remove, inplace=True)

columns_to_rename = {col: col[:-2] for col in filtered_df.columns
                     if col.endswith('_x')}
filtered_df.rename(columns=columns_to_rename, inplace=True)

# Drop observations where 'raised_time' is empty
filtered_df.dropna(subset=['raised_time'], inplace=True)
```

```python
# Merging with Country Stats

df_country = pd.read_csv(country_stats_file, usecols=['country_code',
                                                      'population', 'hdi'])
final_df = filtered_df.merge(df_country, on='country_code', how='left')

# Merge with loan coordinates for later analysis
df_coords = pd.read_csv(loan_coords_file)
final_df = final_df.merge(df_coords, on='loan_id', how='left')

# Function to classify gender dummies
def classify_gender(gender_str):
    if pd.isna(gender_str):
        return np.nan  # Handle missing values
    genders = set(gender_str.lower().replace(' ', '').split(','))
    if genders == {'female'}:
        return 'female'
    elif genders == {'male'}:
        return 'male'
    else:
        return 'mixed'

# Apply the classification to the 'borrower_genders' column
final_df['gender_group'] = final_df['borrower_genders'].apply(classify_gender)

# Convert gender categories to dummy variables for regression
gender_dummies = pd.get_dummies(final_df['gender_group'], 
                                prefix='gender_group', drop_first=True)

# Concatenate dummy variables with the main dataframe
final_df = pd.concat([final_df, gender_dummies], axis=1)

# Drop empty observations for HDI and duplicates
final_df.dropna(subset=['hdi', 'latitude', 'longitude'], inplace=True)
final_df.drop(columns=['activity_name', 'sector_name', 'loan_use', 
                       'currency_policy', 
                        'currency_exchange_coverage_rate', 'currency', 
                        'num_journal_entries', 'num_bulk_entries', 
                        'distribution_model', 'region'], inplace=True)
```

```python
# Creating More Dummies and Variables

final_df['funding_time_days'] = (pd.to_datetime(final_df['raised_time']) - 
 pd.to_datetime(final_df['posted_time'])).dt.total_seconds() / (60 * 60 * 24)
final_df = final_df[final_df['funding_time_days'] > 0]
final_df['log_funding_time'] = np.log(final_df['funding_time_days'])

# Create sector fixed effects (one-hot encoding)
sector_dummies = pd.get_dummies(final_df['sector'], prefix='sector', 
                                drop_first=True)
final_df = pd.concat([final_df, sector_dummies], axis=1)

# Aggregate the number of loans per country for later analysis
loan_counts = final_df.groupby('country_code').size().reset_index(name='loan_count')

# Convert all sector dummy variables from bool to int for later analysis
sector_columns = [col for col in final_df.columns if 
                  col.startswith('sector_')]
final_df[sector_columns] = final_df[sector_columns].astype(int)
```

Conversely, sectors such as entertainment (0.11%), wholesale (0.10%), and construction (0.93%) have minimal representation, indicating a lower demand for loans in these areas.

```python
# Table III: Gender-Based Summary Statistics with Differences

import pandas as pd
from IPython.display import display, HTML

# Function to calculate statistics
def calculate_stats(var):
    male_data = final_df[final_df['gender_group_male'] == 1][var]
    female_data = final_df[(final_df['gender_group_male'] == 0) & (final_df['gender_group_mixed'] == 0)][var]
    mixed_data = final_df[final_df['gender_group_mixed'] == 1][var]
    
    return [
        male_data.mean(), male_data.std(),
        female_data.mean(), female_data.std(),
        mixed_data.mean(), mixed_data.std(),
        male_data.mean() - female_data.mean(), male_data.std() - female_data.std()
    ]

# Define continuous variables and their labels
continuous_vars = ['funding_time_days', 'loan_amount', 'hdi', 'lender_term', 'num_lenders_total']
variable_names = {
    'funding_time_days': 'Funding Time (Days)',
    'loan_amount': 'Loan Amount (USD)',
    'hdi': 'HDI (0-1)',
    'lender_term': 'Loan Term (Months)',
    'num_lenders_total': 'Number of Lenders (Per Loan)'
}

# Generate summary statistics
stats_data = {variable_names[var]: calculate_stats(var) for var in continuous_vars}

# Create DataFrame with MultiIndex column headers
stats_summary = pd.DataFrame.from_dict(stats_data, orient='index',
                                       columns=pd.MultiIndex.from_tuples([
                                           ('Male', 'Mean'), ('Male', 'S.D'),
                                           ('Female', 'Mean'), ('Female', 'S.D'),
                                           ('Mixed', 'Mean'), ('Mixed', 'S.D'),
                                           ('Male - Female', 'Mean'), ('Male - Female', 'S.D')
                                       ]))

# Format numbers to 2 decimal places
stats_summary = stats_summary.map(lambda x: f'{x:.2f}')

# Display Title
display(HTML("<h3 style='text-align: center;'>Table III: Gender-Based Summary Statistics with Differences</h3>"))

# Style table for compact display
styled_table = (
    stats_summary.style
    .set_table_styles([
        {'selector': 'th', 'props': [('text-align', 'center'), ('background-color', 'white'), ('padding', '5px'), ('font-size', '10pt')]},  
        {'selector': 'td', 'props': [('text-align', 'center'), ('background-color', 'white'), ('padding', '5px'), ('font-size', '9pt')]},  
        {'selector': 'table', 'props': [('margin-left', 'auto'), ('margin-right', 'auto'), ('border-collapse', 'collapse'), ('width', '90%')]}
    ])
)

styled_table
```

Table 3 presents gender based summary statistics for key loan characteristics, comparing male, female, and mixed-gender borrower groups. Notably, female borrowers exhibit significantly shorter average funding times of 12.49 days compared to male borrowers 18.20 days, with a mean difference of 5.71 days, suggesting a potential lender preference for funding female-led loans more quickly. Additionally, female borrowers tend to request smaller loan amounts, on average (673.66 USD), relative to male borrowers (784.68 USD), while mixed-gender groups request substantially larger loans (1,699.92 USD), possibly reflecting the collective nature of such loans. Loan terms are also shorter for female borrowers (13.06 months) compared to males (15.88 months), with a mean difference of 2.82 months, indicating that female borrowers may prefer to or be offered more conservative loan durations. Finally, loans to female borrowers attract fewer lenders on average (17.76) compared to male borrowers (23.28), while mixed-gender loans engage significantly more lenders (37.83).

# 4. Plots, Histograms, Figures

```python
# Figure I: Distribution of Loan Variables
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Figure I: Distribution of Loan Variables', fontsize=16, 
             fontweight='bold')

variables = [
    ('num_lenders_total', 'Lender Count', (0, 200), 500),
    ('loan_amount', 'Loan Amount (USD)', (0, 10000), 500),  
    ('term_in_months', 'Loan Terms (Months)', (0, 60), 60), 
    ('funding_time_days', 'Funding Time (Days)', None, 60) 
]

for ax, (var, xlabel, xlim, bins) in zip(axes.flatten(), variables):
    sns.histplot(final_df[var], bins=bins, kde=True, color='dodgerblue', 
                 alpha=0.7, ax=ax)
    ax.set_xlabel(xlabel, fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    ax.set_title(f'Distribution of {xlabel}', fontsize=14)
    if xlim:
        ax.set_xlim(xlim)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

To visualise the distribution of key variables, we plot them on a histogram. These distributions reveal significant right-skewness across all four loan variables. The lender count and loan amount distributions indicate that most loans are relatively small, with a few outliers receiving significantly higher amounts and attracting more lenders. The loan term distribution appears multimodal, suggesting distinct categories of loan durations rather than a continuous spread. Finally, funding time exhibits a peak at very short durations but also a long tail, suggesting that while many loans are funded quickly, a notable portion experience substantial delays.

```python
# Figure II: Factors Affecting Funding Time by Gender
def compute_percentiles(df, x_col, y_col, group_col, num_bins=10):
    df = df[[x_col, y_col, group_col]].dropna()
    
    if df[x_col].nunique() < num_bins:
        print(f"Warning: {x_col} has fewer unique values than bins. Adjusting bin count.")
        num_bins = df[x_col].nunique() 
    
    try:
        df['percentile'] = pd.qcut(df[x_col], q=num_bins, labels=False, 
                                   duplicates='drop')
    except ValueError:
        df['percentile'] = pd.cut(df[x_col], 
                                  bins=np.linspace(df[x_col].min(), 
                                                   df[x_col].max(), num_bins), 
                                  labels=False, include_lowest=True)
    
    return df.groupby(['percentile', group_col])[y_col].mean().reset_index()

avg_funding_by_term = compute_percentiles(final_df, 'term_in_months', 
                                          'funding_time_days', 'gender_group')
avg_funding_by_size = compute_percentiles(final_df, 'loan_amount', 
                                          'funding_time_days', 'gender_group')
avg_funding_by_lender = compute_percentiles(final_df, 'lender_count', 
                                            'funding_time_days', 'gender_group')
avg_funding_by_hdi = compute_percentiles(final_df, 'hdi', 
                                         'funding_time_days', 'gender_group')

color_mapping = {
    'male': 'dodgerblue',
    'mixed': 'orange',
    'female': 'red'
}

fig, axes = plt.subplots(4, 1, figsize=(5, 10))

def plot_funding_time(ax, data, title, xlabel):
    for gender in final_df['gender_group'].unique():
        subset = data[data['gender_group'] == gender]
        if not subset.empty:
            ax.plot(subset['percentile'], subset['funding_time_days'], 
                    label=gender, color=color_mapping.get(gender, 'black'))
    ax.set_title(title)
    ax.set_xlabel(xlabel + ' (percentiles)')
    ax.set_ylabel('Average Funding Time (days)')
    ax.legend()

plot_funding_time(axes[0], avg_funding_by_term, 
                  'Average Funding Time by Loan Term', 'Loan Term')
plot_funding_time(axes[1], avg_funding_by_size, 
                  'Average Funding Time by Loan Size', 'Loan Size')
plot_funding_time(axes[2], avg_funding_by_lender, 
                  'Average Funding Time by Lender Count', 'Lender Count')
plot_funding_time(axes[3], avg_funding_by_hdi, 
                  'Average Funding Time by HDI', 'Human Development Index')

fig.suptitle('Figure II: Factors Affecting Funding Time by Gender', 
             fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

We can now take a close look at the relationship between funding time and loan characteristics, varied by gender. Across loan term, size, and lender count, longer durations and larger loans are associated with increased funding time. Male-led loans generally experience the longest funding times, followed by mixed-gender and female-led loans. This trend is particularly pronounced in larger loans and those with more lenders, suggesting that gender disparities persist even when controlling for key loan factors. The relationship with Human Development Index (HDI) is less clear, indicating that funding times may not follow a clear pattern based on regional economic conditions.

```python
# Figure III: Funding Time Distribution by Gender
plt.figure(figsize=(10, 6))

palette = {"male": "dodgerblue", "female": "red", "mixed": "orange"}

sns.boxplot(
    x="gender_group",
    y="funding_time_days",
    hue="gender_group",
    data=final_df,
    palette=palette,
    legend=False, 
    flierprops={'marker': 'o', 'markerfacecolor': 'grey', 'alpha': 0.1}  
)

plt.xlabel("Gender", fontsize=12)
plt.ylabel("Funding Time (Days)", fontsize=12)
plt.title("Figure III: Funding Time Distribution by Gender", 
          fontsize=16, fontweight="bold")

plt.show()
```

Clearly, there seems to be a relationship between gender and funding speed. We can investigate this further with a box plot illustrating the distribution of funding times across gender groups. Female-led loans have the shortest median funding time, with a tighter interquartile range, indicating more consistent funding speeds. Male-led and mixed-gender loans show a higher median funding time and a wider spread, suggesting greater variability in funding duration. Additionally, male-led loans exhibit more extreme outliers, indicating that some take significantly longer to fund. This pattern suggests potential gender-based differences in the speed at which loans are funded, with female-led loans tending to receive funding more quickly.

```python
# Figure IV: Ranking of Sectors by Mean Funding Time
sector_funding_means = final_df.groupby('sector')['funding_time_days'].mean().sort_values(ascending=True)

plt.figure(figsize=(10, 6))

sns.barplot(x=sector_funding_means.values, y=sector_funding_means.index, 
            hue=sector_funding_means.index, palette='Blues', legend=False)

plt.xlabel('Mean Funding Time (Days)', fontsize=12)
plt.ylabel('Sector', fontsize=12)
plt.title('Figure IV: Ranking of Sectors by Mean Funding Time', 
          fontsize=16, fontweight='bold')

plt.show()
```

Furthermore, we can diffentiate loans based on the sector type. To illustrate sectoral funding differences, we plot the ranking of sectors by mean funding time on a bar chart. Sectors such as arts, manufacturing, and education have the shortest average funding times, suggesting that loans in these categories are funded more quickly. In contrast, transportation, clothing, and services experience the longest funding times, potentially indicating lower lender interest or higher loan amounts requiring more time to secure full funding. The overall trend suggests that funding speed varies significantly across industries, highlighting sector-specific dynamics that influence loan accessibility.

```python
# Figure V: Heatmap of Average Funding Time by Sector and Gender
pivot_df = final_df.pivot_table(index="sector", columns="gender_group",
                                values="funding_time_days", aggfunc="mean")

plt.figure(figsize=(10, 6))

sns.heatmap(pivot_df, annot=True, cmap="coolwarm", linewidths=0.5, 
            cbar=False)

plt.xlabel("Gender")
plt.ylabel("Sector")
plt.title("Figure V: Heatmap of Average Funding Time by Sector and Gender",
          fontsize=16, fontweight="bold")

plt.show()
```

Evidently, there is a clear relationship between funding speed and two key variables: gender and sector. We plot a heatmap to visualise the average funding time across different sectors and gender groups. Warmer colors (red) indicate longer funding times, while cooler colors (blue) represent shorter funding durations. Across most sectors, male-led and mixed-gender groups experience longer funding times compared to female-led groups, which generally receive funding more quickly. Notably, sectors such as housing, retail, and food exhibit some of the longest funding times, particularly for male-led enterprises. In contrast, education and arts show the shortest funding times across all gender groups. This pattern suggests that funding efficiency varies significantly by both sector and gender composition, with some industries and borrower demographics facing greater challenges in securing quick funding.

# 5. Conclusion

This study highlights key factors influencing microloan funding speed, with gender and sector emerging as the most significant determinants. Female-led loans tend to receive funding more quickly, suggesting a potential lender preference or institutional prioritisation of womens financial inclusion. Sectoral differences also play a crucial role, with industries such as arts, education, and manufacturing receiving faster funding, while transportation, clothing, and services experience longer wait times.

Loan characteristics, including size and term length, further impact funding speed, with larger and longer-term loans facing delays. However, macroeconomic indicators like HDI show no clear relationship with funding time, indicating that borrower-specific and loan-specific factors outweigh broader economic conditions in influencing lender behavior.
